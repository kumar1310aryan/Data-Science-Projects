# About Dataset

The dataset includes 244 instances that regroup a data of two regions of Algeria, namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012.
The dataset includes 11 attributes and 1 output attribute (class)
The 244 instances have been classified into fire (138 classes) and not fire (106 classes) classes.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_csv("Algerian_forest_fires_dataset.csv")
```

# Data Cleaning

```python
## missing values
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.head(125)
```

The dataset is converted into two sets based on the Region from 122th index, we can make a new column based on the region:

1. "Bejaia Region Dataset"
2. "Sidi-Bel Abbes Region Dataset"

Add new column with region

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df=dataset
```

```python
df.info()
```

```python
df.head()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.head(125)
```

```python
df.isnull().sum()
```

### Removing the null valules

```python
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.head(124)
```

```python
df=df.drop(df.index[122]).reset_index(drop=True)
```

```python
df.head(123)
```

```python
df
```

From above its clearly visible

1. 0-121 ("Bejaia Region Dataset") Region=0
2. (122-241)"Sidi-Bel Abbes Region Dataset") Region=1

### Fix spaces in columns

```python
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

### Changing the required columns as integer data type

```python
df.columns
```

```python
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
df.head()
```

### Changing the other columns to float datatype

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
objects
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

### Save the cleaned dataset

```python
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```
